 ** Merging and Labelling Resistant and Non-resistant Proteins **

```python
from google.colab import files
import random




resistant_file = "/content/cleaned_resistant ALL.fasta"
non_resistant_file = "/content/cleaned_Nonresistant.fasta"

output_file = "merged_labeled.fasta"

# =============Read FASTA and Add Labels====================

def read_fasta_with_label(input_file, label):
    sequences = []
    header = ""
    seq = ""

    with open(input_file, "r") as infile:
        for line in infile:
            line = line.strip()

            if line.startswith(">"):
                if header:
                    sequences.append((header, seq))

                # Add label to the header
                header = line + f" | label={label}"
                seq = ""

            else:
                seq += line

        # Save the last sequence
        if header:
            sequences.append((header, seq))

    return sequences

# ===============================Read Both Files==================================

print("Reading resistant sequences...")
resistant = read_fasta_with_label(resistant_file, 1)

print("Reading non-resistant sequences...")
non_resistant = read_fasta_with_label(non_resistant_file, 0)

# ======== Merge and shuffle =========

all_sequences = resistant + non_resistant
random.shuffle(all_sequences)

print(f"\nTotal Resistant Sequences     : {len(resistant)}")
print(f"Total Non-resistant Sequences : {len(non_resistant)}")
print(f"Total Sequences               : {len(all_sequences)}")


# =======================Saving Merged and Shuffled file into  FASTA===========================

with open(output_file, "w") as outfile:
    for header, seq in all_sequences:
        outfile.write(header + "\n")
        outfile.write(seq + "\n")

print("\n File has been merged, labeled and shuffled!")
files.download(output_file)
print(f'Output save to : {output_file}')
```

**Parse and Split Data into Train and Test **

```python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from sklearn.model_selection import train_test_split
import numpy as np


uploaded = files.upload()
file_name = "merged_labeled.fasta"

sequences = []
labels_list = []

for record in SeqIO.parse(file_name, "fasta"):
    sequences.append(str(record.seq))

    if "label=1" in record.description:
        labels_list.append(1)
    else:
        labels_list.append(0)

print("========== DATA SUMMARY ==========")
print(f"Total sequences      : {len(sequences)}")
print(f"Resistant            : {labels_list.count(1)}")
print(f"Non-resistant        : {labels_list.count(0)}")

# ===========Split the dataset==========

X_train_raw, X_test_raw, Y_train, Y_test = train_test_split(
    sequences,
    labels_list,
    test_size=0.20,
    stratify=labels_list,
    random_state=42
)

print("\n========== TRAIN TEST SPLIT ==========")
print(f"Training set size    : {len(X_train_raw)}")
print(f"Testing set size     : {len(X_test_raw)}")

# =========Count classes in training set =========

Y_train_np = np.array(Y_train)

resistant_count = np.sum(Y_train_np == 1)
non_resistant_count = np.sum(Y_train_np == 0)

print("\n========== TRAINING SET ==========")
print(f"Resistant            : {resistant_count}")
print(f"Non-resistant        : {non_resistant_count}")

# ===========Save training and testing FASTA files============


def save_to_fasta(sequences, labels, filename):
    records = []

    for i, (seq, label) in enumerate(zip(sequences, labels)):
        records.append(
            SeqRecord(
                Seq(seq),
                id=f"seq_{i}",
                description=f"split_data | label={label}"
            )
        )


    with open(filename, "w") as output_handle:
        SeqIO.write(records, output_handle, "fasta")

    print(f"Saved {len(records)} sequences to: {filename}")


save_to_fasta(X_train_raw, Y_train, "train_data.fasta")
save_to_fasta(X_test_raw, Y_test, "test_data.fasta")
files.download("train_data.fasta")
files.download("test_data.fasta")
```

              ** ProtBERT Feature Extraction**

```python
import torch
from transformers import BertModel, BertTokenizer
from Bio import SeqIO
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Model-loading
tokenizer = BertTokenizer.from_pretrained("Rostlab/prot_bert", do_lower_case=False)
model = BertModel.from_pretrained("Rostlab/prot_bert")
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
model.eval()

def get_embeddings(filename):

    sequences = []
    labels = []

    # Read sequences and labels
    for record in SeqIO.parse(filename, "fasta"):

        seq = " ".join(list(str(record.seq)))
        sequences.append(seq)

        if "label=1" in record.description:
            labels.append(1)
        else:
            labels.append(0)

    # Extract ProtBERT embeddings
    all_embeddings = []

    with torch.no_grad():

        for seq in sequences:

            ids = tokenizer(
                seq,
                return_tensors="pt",
                padding=True,
                truncation=True,
                max_length=512
            ).to(device)

            output = model(**ids)

            embedding = output.last_hidden_state.mean(dim=1).squeeze().cpu().numpy()

            all_embeddings.append(embedding)

    return np.array(all_embeddings), np.array(labels)

# FEATURE EXTRACTION
print("Extracting training features...")
X_train, y_train = get_embeddings("/content/train_data.fasta")

print("Extracting testing features...")
X_test, y_test = get_embeddings("/content/test_data.fasta")

print("Feature shapes:", X_train.shape, X_test.shape)
```

** Initializing Random Forest with balanced class weight **

```python
#clf is random forest classifier
#initialize classifier with 100 decision trees
clf = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
print("Training the Random Forest with balanced weights")
clf.fit(X_train, y_train)

# making predictions
print("Predicting on test data...")
y_pred = clf.predict(X_test)

# Results
print("\n--- Classification Report ---")
print(classification_report(y_test, y_pred))

print("\n--- Confusion Matrix ---")
print(confusion_matrix(y_test, y_pred))
```

** Virtual Mutation Analysis **

```python
import numpy as np
import torch

def virtual_mutation_test(sequence, pos_to_mutate, new_amino_acid):
    """
    sequence: The original protein string (e.g., 'MAVK...')
    pos_to_mutate: List of indices to change
    new_amino_acid: The AA to insert (e.g., 'T' for Threonine)
    """

    #  Create the mutated sequence
    seq_list = list(sequence.replace(" ", "")) # remove spaces if present
    for pos in pos_to_mutate:
        if pos < len(seq_list):
            seq_list[pos] = new_amino_acid

    mutated_seq = " ".join(seq_list)
    original_seq = " ".join(list(sequence.replace(" ", "")))

    #  Extract embeddings for both
    def get_single_embedding(s):
        ids = tokenizer(s, return_tensors='pt', padding=True, truncation=True, max_length=512).to(device)
        with torch.no_grad():
            output = model(**ids)
        return output.last_hidden_state.mean(dim=1).squeeze().cpu().numpy().reshape(1, -1)

    orig_emb = get_single_embedding(original_seq)
    mut_emb = get_single_embedding(mutated_seq)

    #  Predict probabilities
    orig_prob = clf.predict_proba(orig_emb)[0][1]
    mut_prob = clf.predict_proba(mut_emb)[0][1]

    print(f"--- Mutation Results ({new_amino_acid}) ---")
    print(f"Original Resistance Probability: {orig_prob:.4f}")
    print(f"Mutated Resistance Probability:  {mut_prob:.4f}")
    print(f"Change: {((mut_prob - orig_prob) / orig_prob * 100):.2f}%")

# --- EXECUTION ---
# Pick a sequence from  test set that is currently labeled 'Non-Resistant'
sample_idx = 13
sample_seq = X_test_raw[sample_idx]

# mutate 5 random positions to Threonine
positions = list(range(0, 50))
virtual_mutation_test(sample_seq, positions, 'T')
```
